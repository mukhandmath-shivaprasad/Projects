```python
grouped = df.groupby(by='State')
grouped.first()
```

```python
# Get group for the state of Karnataka and assign to variable kar_state

kar_state =  grouped.get_group('Karnataka')
```

```python
# Reset index to new values by dropping old indices
kar_state.reset_index(drop= True)
```

There are 30 districts in Karnataka States as per 2011 census data (presentlt there are 31 districts)

```python
kar_state = kar_state.drop(['Sub-districts', 'State'], axis =1)
```

```python
# Stats of the State of Karnataka
area = kar_state['Area (km2)'].sum(axis = 0)
print(f'Total area of Karnataka state is : {area} sq.km')
print(" ")
population = kar_state['Population'].sum(axis = 0)
print(f'Total population of Karnataka state is {population}')
print(" ")
lit_rate = kar_state['Literacy'].mean()
print(f'Average literacy rate of Karnataka is {lit_rate:.2f}%')
print(" ")
area_of_India =df['Area (km2)'].sum(axis = 0)
pop_of_India = df['Population'].sum(axis = 0)
prct_area = (area / area_of_India)*100
prct_pop = (population / pop_of_India)*100
print(f"Karnataka is {prct_area:.2f}% of total area and {prct_pop:.2f}% of total population of India ")
print(" ")
```
